# Exploratory Fraud Analysis

This notebook provides quick exploratory checks on processed PaySim transactions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/processed/clean_transactions.csv')
df.head()
```

```python
fraud_rate = (df['isFraud'].sum() / len(df)) * 100
print(f'Fraud rate: {fraud_rate:.4f}%')
```

```python
hourly = df[df['isFraud'] == 1].groupby('hour').size().reset_index(name='fraud_count')
sns.barplot(data=hourly, x='hour', y='fraud_count', color='steelblue')
plt.title('Fraud Transactions by Hour')
plt.tight_layout()
plt.show()
```
